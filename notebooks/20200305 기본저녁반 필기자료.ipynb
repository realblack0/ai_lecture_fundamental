### url로 데이터 불러오기

`read_csv`는 웹에서 url로 데이터를 받아오는 것도 가능하다.  
uci는 기계학습 데이터 공개 저장소이다. 기계학습의 대표적인 연습용 데이터셋인 iris 데이터를 받아오자.

iris 데이터 저장소 링크:
https://archive.ics.uci.edu/ml/machine-learning-databases/iris/

```python
import pandas as pd
```

```python
# header=None옵션을 주지 않으면 첫번째 행을 자동으로 열 이름으로 사용한다.
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header=None)
```

```python
data.head()
```

행과 열로 구성되어 있고 표 형태로 생긴 것은 DataFrame 객체이다.

```python
type(data)
```

시작하기 전에 데이터 원본 보존용으로 카피해둔다.

```python
data_raw = data.copy() # pandas의 copy는 deep copy
```

### 이름 바꾸기

`rename`은 이름을 바꿀 수 있다.  
{ 원래이름 : 바꿀이름 }으로 된 딕셔너리와, 축(axis)을 지정한다.

```python
data.rename({0:"Sepal Length", 1:"Sepal Width"}, axis=1) # axis=1 이므로 열 이름 변경
```

`rename`은 결과값은 있지만 자기자신을 바꾸지 않는다.

```python
data.head() # 바뀌지 않았음
```

결과를 반영할 때는 재할당한다.

```python
data = data.rename({0:"Sepal Length", 1:"Sepal Width"}, axis=1)  # 재할당
```

```python
data.head()
```

또는 `inplace`옵션을 이용해서 mutable처럼 자기 자신을 바꿀 수 있다.

```python
data = data_raw.copy() # 원본 데이터 복원
```

```python
data.rename({0:"Sepal Length", 1:"Sepal Width"}, axis=1, inplace=True)  # mutable처럼 자기 자신 바꿈
                                                                        # return이 없다.
```

```python
data.head() # 자기 자신을 바꿈
```

바꾼 이름으로 데이터를 뽑아올 수 있다.

```python
data["Sepal Width"]
```

`axis=1` 대신에 키워드 파라미터 `columns`를 이용할 수 있다.

```python
data = data_raw.copy() # 원본 데이터 복원
```

```python
data = data.rename(columns=({0:"Sepal Length", 1:"Sepal Width"}))
```

```python
data.head()
```

`axis=0`은 인덱스 이름을 바꾼다. (`axis=0`은 행 방향)

```python
data2 = data.head()
```

```python
data2.rename({0:"aa", 1:"bb", 2:"cc", 3:"dd"}, 
             axis=0, # axis의 default value는 0
             inplace=True) 
```

```python
data2
```

키워드 파라미터 `index`를 이용해서 인덱스 이름을 바꿀 수도 있다

```python
data2.rename(index={4:"ee"})
```

`read_csv`로 데이터를 불러올 때부터 이름을 지정하는 것도 가능하다. (`names` 옵션을 이용)

## 원하는 위치의 값 뽑기

### loc과 iloc

**loc\[ \]**
- loc는 이름으로 뽑는다.
- 인덱서이므로 소괄호( ) 대신 대괄호 \[ \]를 사용한다.  

**iloc\[ \]**
- iloc는 인덱스로 뽑는다.  
- 인덱서이므로 소괄호( ) 대신 대괄호 \[ \]를 사용한다. 

```python
data.loc[1]
```

```python
data.loc[1] # 인덱스 이름이 1인 행을 뽑음
```

```python
data.iloc[1] # 인덱스가 1인 행을 뽑음
```

`loc`은 인덱스 이름으로 접근하기 때문에 인덱스 이름을 바꾼 data2에서는 숫자로 접근할 수 없다.

```python
data2
```

```python
# 인덱스로 접근 불가, 실행하면 Error
data2.loc[1]

# KeyError: 1
```

```python
data2.loc['bb'] # 인덱스 이름으로 접근
```

`iloc`는 인덱스로 접근하기 때문에 인덱스 이름과 상관이 없다.

```python
data2.iloc[1]
```

`iloc`와 `loc` 모두 numpy 인덱싱 처럼 ( : ) 를 사용할 수 있다.  
단, loc는 이름을 기반으로 접근하기 때문에 마지막 행을 포함한다.

```python
data.iloc[0:3] # 0번째~2번째 행 슬라이싱
               # 3번째 행 미포함
```

```python
data.loc[0:3] # 이름이 0인 행 ~ 이름이 3인 행 슬라이싱
              # 이름이 3인 행 포함
```

두번째 ( : )을 이용해서 간격을 지정할 수 있다.

```python
data.iloc[::2] # 모든 행을 2칸 간격으로 슬라이싱
```

### boolean 인덱싱

비교 연산자를 이용해서 dtype이 bool인 Series를 객체를 만들 수 있다.

```python
mask = data.b > 4
mask
```

`loc`에서 dtype이 bool인 Series로 boolean 인덱싱을 할 수 있다.

```python
data.loc[mask] # mask의 값이 True인 행만 인덱싱
               # b열의 값이 4보다 큰 행만 인덱싱
```

bool 타입을 반환하는 lambda 함수를 이용해서 boolean 인덱싱을 할 수도 있다.

```python
data.loc[lambda row:row.b > 4] # lambda 함수에는 행이 하나씩 인자로 주어짐.
```

괄호안에 lambda함수만 있는 경우에는 DataFrame에서 boolean 인덱싱하는 것과 결과가 같다.

```python
data[lambda row:row.b > 4]
```

`loc`은 boolean 인덱싱 결과에서, 뽑을 열을 지정할 수 있다.

```python
data.loc[lambda row:row.b > 4, ['a', 'b']] # 결과에서 a, b만 가져옴
```

`lambda` 대신에 함수를 쓸 수 있다. 함수를 이용하면 더 복잡한 조건을 만들 수 있다.

```python
def x(df):
    return df.b > 4

data.loc[x]
```

괄호 안에 조건을 넣을 수도 있다.

```python
data.loc[data.b > 4]
```

복잡한 조건식도 넣을 수 있다. 이때, 조건은 괄호로 묶어야 한다.

```python
# a열 값이 2보다 크고, c열 값도 2보다 큰 행들에서 b열의 값만 뽑음
data.loc[(data.a>2) & (data.c>2), "b"] 
```

`iloc`도 boolean 인덱싱이 가능하다. 단, `iloc`에서는 boolean 인덱싱할 때는 indexable 객체를 mask로 사용할 수 없다.  
Series와 DataFrame 객체는 indexable이므로 위와 같은 방식을 이용하면 Error가 발생한다.

```python
# 실행하면 Error
data.iloc[data.b > 4]

# NotImplementedError: iLocation based boolean indexing on an integer type is not available
```

numpy array는 indexable이 아니므로 array로 변환해서 사용할 수 있다.

```python
mask = data.b > 4
mask = mask.values  # numpy array로 변환 / array는 indexable이 아님.
data.iloc[mask]
```

`iloc`은 이름 대신 열 인덱스를 이용해서 가져올 열을 지정할 수 있다.

```python
data.iloc[mask, [0, 1, 2]] # 열 인덱스가 0, 1, 2인 열만 가져옴
```

### at과 iat

- at과 iat은 값을 1개만 뽑는다.
- loc과 iloc은 값을 여러개 뽑을 수 있다.
- at은 이름으로 접근한다.
- iat은 인덱스로 접근한다.

```python
data2
```

```python
data2.at["aa", 'Sepal Length'] = 100
```

```python
data2.iat[0, 0]
```

```python
data2
```

이외에도 값을 뽑는 방법에는 `ix`이 있었는데, deprecated되었기 때문에 최신 버전에서는 사용할 수 없다.

```python
# deprecated-> 최신버전은 Error
data.ix([0, 2], 'a')
```

`at`, `iat`, `loc`, `iloc`의 속도를 비교해보면 at, iat이 더 빠르다.

```python
%%timeit
data.at[0, 'a']
```

```python
%%timeit
data.iat[0, 0]
```

```python
%%timeit
data.loc[0, 'a']
```

```python
%%timeit
data.iloc[0, 0]
```

## 인덱스 조작

### set_index

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

`set_index`는 특정 열을 인덱스로 만든다.

```python
# tip을 인덱스로 만든다.
tips.set_index('sex')
```

set_index를 사용하면 인덱스 값이 같은 경우가 생길 수 있으므로 조심해야한다.  
인덱스가 같으면 실수 가능성이 생긴다.

```python
# 이름이 같은 것을 다 가져온다.
tips.set_index('sex').loc["Female"]
```

인덱스 이름이 중복될 때에는 `iloc`을 이용해서 원하는 값을 뽑을 수 있다.

```python
# 2번째 여자에 대한 정보
tips.set_index('sex').loc["Female"].iloc[2]
```

`drop=False` 옵션을 이용하면  
원래 있던 sex 열은 그대로 있고, 인덱스가 sex로 새로 생긴다.

```python
tips.set_index('sex', drop=False)
```

`drop=True` 옵션을 이용하면  
데이터에서 sex 열이 없어지고 index만 sex로 생긴다.

```python
tips.set_index('sex', drop=True)
```

### 멀티 인덱스

`set_index`에 2개 이상의 열 이름을 리스트로 묶어서 넣으면 멀티 인덱스가 된다.

```python
tips.set_index(['sex', 'day'])
```

`inplace=True` 옵션을 이용하면 바뀐 값이 적용된다.

```python
tips.set_index(['sex', 'day'], inplace=True)
```

`.index`이 MultiIndex 객체를 반환한다.

```python
tips.index
```

멀티 인덱스로 지정한 열의 값들이 묶여 있다.

```python
tips.index[0]
```

```python
tips.index[0][1]
```

`loc`를 이용해서 이름으로 Multiindex에 접근할 수 있다.

```python
tips.loc['Male']
```

다시 `loc`으로 이름으로 접근해서 값을 가져온다.

```python
tips.loc['Male'].loc['Sun'] # 남자이고, 일요일에 방문한 손님
```

묶어서 넣으면 한번에 가져올 수 있다.

```python
tips.loc[('Male', 'Sun')]
```

### reset_index

멀티인덱스 상태에서 인덱스를 바꿔보자.  
`set_index`로 바꾸면 원래 index였던 sex, day가 사라지는 문제가 있다.

```python
tips.set_index('smoker')
```

`reset_index`를 사용하면 원래 인덱스 모양으로 돌아온다.

```python
tips.reset_index()
```

멀티인덱스에서 초기화할 인덱스 번호를 지정할 수 있다.

```python
tips.reset_index(0) # sex만 초기화
```

```python
tips.reset_index(1) # day만 초기화
```

```python
tips.reset_index([0,1]) # sex, day 모두 초기화
```

### 중복데이터 처리

`duplicated` 메소드는 중복데이터 여부를 True, False로 반환한다.  
조건식을 이용해서 중복 데이터만 뽑는다.

```python
data[data.duplicated()==True]
```

찾아낸 중복 데이터를 `drop`으로 삭제할 수 있다.

```python
data.drop([34, 37, 142]) # 기본값: axis=0 
```

열을 삭제할 때는 axis를 지정해준다.

```python
data.drop('petal_length', axis=1)
```

`drop_duplicates`는 중복 데이터를 찾아서 삭제한다.

```python
data.drop_duplicates()
```
